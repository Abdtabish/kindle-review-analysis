```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(
    'dataset/all_kindle_review .csv',
    encoding='latin-1'
)
data
```

```python
df=data[['reviewText','rating']]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['rating'].value_counts()
```

```python
df['rating']=df['rating'].apply(lambda x: 0 if x<3 else 1)
```

```python
df['reviewText']=df['reviewText'].str.lower()
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from bs4 import BeautifulSoup
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9-]+','',x))
df['reviewText']=df['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english') ]))
df['reviewText']=df['reviewText'].apply(lambda x: re.sub('[http\S+|www\.\S+|\b[\w\.-]+@[\w\.-]+\.\w+\b|[^a-zA-Z0-9 ]','',x))
df['reviewText']=df['reviewText'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())
df['reviewText']=df['reviewText'].apply(lambda x:" ".join(x.split()))

```

```python
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer=WordNetLemmatizer()
```

```python
def lemmatize(sentence):
    return " ".join([lemmatizer.lemmatize(word) for word in sentence.split()])
```

```python
df['reviewText']=df['reviewText'].apply(lambda x: lemmatize(x))
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
X_train_bow=bow.fit_transform(X_train).toarray()
X_test_bow=bow.transform(X_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
X_train_tfidf=tfidf.fit_transform(X_train)
X_test_tfidf=tfidf.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
bow_gnb=MultinomialNB().fit(X_train_bow,y_train)
tfidf_gnb=MultinomialNB().fit(X_train_tfidf,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
y_pred_bow=bow_gnb.predict(X_test_bow)
y_pred_tfidf=tfidf_gnb.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test,y_pred_bow))
```

```python
print(accuracy_score(y_test,y_pred_tfidf))
```

```python
print(confusion_matrix(y_test,y_pred_bow))
```

```python
print(confusion_matrix(y_test,y_pred_tfidf))
```

```python
print(classification_report(y_test,y_pred_bow))
```
